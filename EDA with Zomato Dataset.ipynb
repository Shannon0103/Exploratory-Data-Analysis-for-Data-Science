```python
final_df.columns
```

```python
ratings=final_df.groupby(['Aggregate rating','Rating color','Rating text']).size().reset_index().rename(columns={0:'Rating Count'})
```

```python
ratings
```

Observation
1. When Ratings is between 4.5 to 4.9 --> Excellent
2. When Ratings is between 4.0 to 4.4 --> Very Good
3. When Ratings is between 3.5 to 3.9 --> Good
4. When Ratings is between 2.5 to 3.4 --> Average
5. When Ratings is between 1.8 to 2.4 --> Poor
6. When Ratings is 0 --> Its not Rated

```python
ratings.head()
```

```python
plt.figure(figsize=(12,6))
sns.barplot(x="Aggregate rating",y="Rating Count",data=ratings,hue='Rating color',palette=['white','red','orange','yellow','lightgreen','green'],edgecolor='black')
plt.show()
```

Observations:
1. Count is very high for unrated Category!
2. Maximum number of ratings fall in the average category between 2.5 to 3.4

```python
sns.countplot(x="Rating color",hue="Rating color",data=ratings,palette=['white','red','orange','yellow','lightgreen','green'],edgecolor='black')
plt.show()
#count of ratings
```
